```python
import numpy as np
import pandas as pd
precip=pd.read_csv('data/precip.csv')
```

```python
precip
```

```python
precip=precip.rename(columns={"Tipo-Zone":"zone","Id Sensore": "id", "Data-Ora": "data", "Valore Cumulato":"valore"})
```

```python
del precip['_id']
```

```python
precip
```

```python
rainy = (precip.valore > 0.0)
```

```python
rainy
```

```python
print('media tutti:  ', np.mean(precip.valore))
print('mediana tutti:  ', np.median(precip.valore))
print('max tutti:  ', np.max(precip.valore))
print('min tutti:  ', np.min(precip.valore))
```

```python
print('media tutti:  ', np.mean(precip.valore[rainy]))
print('mediana tutti:  ', np.median(precip.valore[rainy]))
print('max tutti:  ', np.max(precip.valore[rainy]))
print('min tutti:  ', np.min(precip.valore[rainy]))
```
