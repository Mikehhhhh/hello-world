```python
import numpy as np
import pandas as pd
data=pd.read_csv('data/Person_Gender_Height_Weight_Index.csv')
```

```python
bmi_ds=pd.DataFrame(data, columns=['Gender', 'Height', 'Weight'])
```

```python
bmi_ds
```

```python
bmi_ds['BMI']=bmi_ds.Weight/(bmi_ds.Height/100)**2
```

```python
bmi_ds
```

```python
bmi_ds.to_csv('data/Es2_Bmi.csv')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns;
```

```python
sns.histplot(data=bmi_ds, x='BMI', hue='Gender', binwidth=1, kde=True)
```

```python
g=sns.FacetGrid(data=bmi_ds, col='Gender', hue='Gender')
g.map(sns.histplot, 'BMI')
```

```python
sns.scatterplot(data=bmi_ds, x='Weight', y='Height', hue='Gender', style='Gender')
plt.show()
```
