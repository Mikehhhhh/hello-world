```python
import numpy as np
import pandas as pd
```

```python
popolazione = pd.read_csv('data/popolazioneISTAT-regioni1gen2020.csv', sep=';')
popolazione
```

```python
regioni = pd.read_csv('data/italy_regions.csv', sep=';', usecols=['regione', 'superficie'])
regioni
```

```python
regioni.superficie = regioni['superficie']/1000
regioni.superficie.head()
```

```python
popolazione['Territorio'] = popolazione['Territorio'].replace('TrentinoAlto Adige', 'Trentino-Alto Adige')
```

```python
ita = pd.merge(regioni, popolazione, left_on='regione', right_on='Territorio', how='inner').drop('Territorio', axis=1)
ita
```

```python
ita['densità']= ita['Totale']/ita['superficie']
```
